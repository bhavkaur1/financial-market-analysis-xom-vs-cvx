# Stock: XOM (EXXON MOBIL)

```python
xom = yf.Ticker("XOM")
#xom: a remote handle to ExxonMobil’s market data
```

```python
xom_df = xom.history(period="5y")
#Think of this as: “Go to Yahoo Finance, fetch 5 years of ExxonMobil data, and keep it in memory.”
```

```python
xom_df.head()
```

Yahoo Finance (via yfinance) does not always provide an explicit "Adjusted Close" column.
Instead, it provides raw OHLC price data along with separate columns for Dividends and Stock Splits.

This design gives more analytical control:
- Raw prices show actual market trading behavior
- Dividends and splits are tracked explicitly
- Returns can be adjusted manually without double-counting

```python
xom_df['Daily Return'] = xom_df['Close'].pct_change()
```

```python
xom_df.head()
```

```python
#combining it all here to show the updated plot with label, title & data in it
xom_df['Close'].plot()
plt.title('Exxon Mobil (XOM) Closing Price Over Time')
plt.ylabel('Price (USD)')
plt.show()
```

XOM shows a long-term upward price trend with noticeable short-term volatility, indicating growth accompanied by periodic market-driven fluctuations.

```python
xom_df['Daily Return'].plot()
plt.title('Daily Returns of Exxon Mobil (XOM)')
plt.ylabel('Return')
plt.show()
```

```python
xom_df['Daily Return'].hist(bins=50)
plt.title('XOM Distribution of Daily Returns')
plt.ylabel('Frequency')
plt.show()
```

Most XOM daily returns are small
→ Normal day-to-day movement is limited

Large gains or losses are rare
→ Extreme events exist but are infrequent

Center near 0
→ Over one day, XOM neither strongly gains nor loses on average

# Stock: CVX (CHEVRON CORP)


```python
cvx_df = yf.Ticker('CVX').history(period="5y")
```

```python
cvx_df.head()
```

```python
cvx_df['Close'].plot()
plt.title('Chevron (CVX) Closing Price OVer Time')
plt.ylabel('Price (USD)')
plt.show()
```

Chevron displayed a sharp, sudden rise within the year 2022. Definitely related to a strong economic event, or rapid growth of the company fundamentals. And after that too, it has shown a steady growth, with some dips. But it has earned place as a trusted stock among the market.

```python
#calculating Daily Returns for CVX
cvx_df['Daily Return'] = cvx_df['Close'].pct_change()
```

```python
cvx_df.head()
```

```python
#plotting daily returns now
cvx_df['Daily Return'].plot()
plt.title('Daily Returns of Chevron (CVX)')
plt.ylabel('Return')
plt.show()
```

```python
#now plotting Daily Returns in Histogram for analysing Daily Returns against the Frequency of days
cvx_df['Daily Return'].hist(bins=50)
plt.title('CVX Distribution of Daily Returns')
plt.ylabel('Frequency')
plt.show()
```

The daily returns histogram for CVX is centered close to 0%, forming a roughly bell-shaped distribution. This indicates that most daily price changes are small, with positive and negative returns occurring around a stable average. Occasional larger returns on either side reflect periods of heightened volatility, likely driven by major market or macroeconomic events (such as the sharp price movement observed in 2022). Overall, the distribution suggests a relatively stable stock with episodic shocks rather than persistent extreme volatility.



---

# Volatility Comparison

```python
xom_vlt = xom_df['Daily Return'].std()
xom_vlt = round(xom_vlt * 100, 2)
print ("XOM Volatility: ", xom_vlt, "%")
```

```python
cvx_vlt = cvx_df['Daily Return'].std()
cvx_vlt = round(cvx_vlt * 100, 2)
print ("CVX Volatility: ", cvx_vlt, "%")
```

Both stocks show a similar risk–return profile. XOM offers slightly higher average daily returns, but also slightly higher volatility. CVX appears marginally more stable, suggesting a more conservative risk profile. Neither stock clearly dominates without a formal risk-adjusted metric.



---

# Rewards (Returns) Comparison

```python
xom_mr = xom_df['Daily Return'].mean()
print(xom_mr)
print(" -> XOM Mean Daily Return: ", f"{xom_mr: .3%}")
```

```python
cvx_mr = cvx_df['Daily Return'].mean()
print(cvx_mr)
print(" -> CVX Mean Daily Return: ", f"{cvx_mr: .3%}")
```

*XOM exhibits a slightly higher average daily return than CVX, accompanied by marginally higher volatility, while CVX offers a more stable return profile with lower daily fluctuations.*



---


# Risk-Adjusted Returns

```python
xom_rar = xom_mr / xom_vlt
cvx_rar = cvx_mr / cvx_vlt
print ("XOM Risk Adjusted Return: ", f"{xom_rar: .3%}")
print ("CVX Risk Adjusted Return: ", f"{cvx_rar: .3%}")
```

# Conclusion

**Although CVX exhibits lower daily volatility, XOM delivers superior risk-adjusted returns, indicating more efficient compensation for risk over the analyzed period.**
